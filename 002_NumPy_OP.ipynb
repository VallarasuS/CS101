## Operations

Arithmetic operations on NumPy arrays are applied element wise, and returns a 'NEW' array with results.


```python
import numpy as np

a = np.array([10,20,30,40])
b = a + 1
print(a)
print(b)

c = b - a
print(c)
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 2, 3, 4])

z = x @ y
print(z)
```

Some short hand operations like `*=` , `+=` modify the source array in place rather than creating a new one


```python
a_ones = np.ones((2,3))
print(a_ones)

a_ones += 1
print(a_ones)
```

```python
scores = np.array([50,45,30,49,62])
results = scores > 35

print("Pass / Faile ? :", results)
print("Centum", scores == 0)
```
